```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

image = cv2.imread ('/download.jpg')

image_rgb = cv2.cvtColor(image ,cv2.COLOR_BGR2RGB)

plt.imshow(image_rgb)
plt.axis('off')
plt.show()
```

```python
def adjust_brightness_contrest(image, brightness=0, contrast=0):
   return cv2.convertScaleAbs(image, alpha=1 + contrast / 100, beta=brightness)
image_adjusted = adjust_brightness_contrest(image, brightness=10, contrast=60)
plt.imshow(cv2.cvtColor(image_adjusted, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()

```

```python
plt.hist(image_rgb.ravel(), bins=8, range=[0, 256])
plt.title('Histogramme ')
plt.xlabel('Instensité de pixels')
plt.ylabel('Nombre de pixels')
plt.show()
```

```python
blurred_image = cv2.GaussianBlur(image_adjusted, (7, 5), 8)
plt.imshow(cv2.cvtColor(blurred_image,  cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python


sobelex = cv2.Sobel(image_adjusted, cv2.CV_64F, 1, 0, ksize=5)
sobeley = cv2.Sobel(image_adjusted, cv2.CV_64F,0,1,ksize=5)
abs_soblex=cv2.convertScaleAbs(sobelex)
abs_sobley=cv2.convertScaleAbs(sobeley)

plt.subplot(1,2,1),plt.imshow(sobelex,cmap='gray')
plt.title('solex X')
plt.axis('off')

plt.subplot(1,2,2),plt.imshow(sobelex,cmap='gray')
plt.title('solex y')
plt.axis('off')
```

```python
# Get image dimensions (for RGB image, shape returns (rows, cols, channels))
rows, cols = image_rgb.shape[:2]  # Only get first two dimensions

# Create rotation matrix (center, angle, scale)
rotation_matrix = cv2.getRotationMatrix2D((cols / 2, rows / 2), 45, 1)

# Apply rotation to the RGB image (not gray_image)
rotated_image = cv2.warpAffine(image_rgb, rotation_matrix, (cols, rows))  # Fixed: warpAffine not warpAffline

plt.imshow(rotated_image)  # Removed cmap='gray' since it's RGB
plt.axis('off')
plt.show()
```

```python
# Get image dimensions (for RGB image, shape returns (rows, cols, channels))
rows, cols = image_adjusted.shape[:2]  # Use image_adjusted instead of image_rgb

# Create rotation matrix (center, angle, scale)
rotation_matrix = cv2.getRotationMatrix2D((cols / 2, rows / 2), 45, 1)

# Apply rotation to the adjusted image (convert to RGB for display)
rotated_image = cv2.warpAffine(image_adjusted, rotation_matrix, (cols, rows))  # Fixed: warpAffine not warpAffline

# Convert BGR to RGB for proper display
rotated_image_rgb = cv2.cvtColor(rotated_image, cv2.COLOR_BGR2RGB)

plt.imshow(rotated_image_rgb)  # Removed cmap='gray' since it's color
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np

image = cv2.imread('download.jpg')
mean = 0
sigma = 20
g_noise = np.random.normal(mean, sigma, image.shape)
noisy_image = cv2.add(image, g_noise.astype(np.uint8))  # Fixed: added cv2.add and proper casting

# Convert to RGB for display
noisy_image_rgb = cv2.cvtColor(noisy_image, cv2.COLOR_BGR2RGB)

plt.imshow(noisy_image_rgb)
plt.axis('off')
plt.show()
```
